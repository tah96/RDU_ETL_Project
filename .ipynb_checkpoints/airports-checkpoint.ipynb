```python
import pandas as pd
```

```python
#Define file path

file = "Original_CSVs/airports.csv"
```

```python
#Read CSV into Pandas

df = pd.read_csv(file)
```

```python
df.head()
```

```python
#Drop null fields

drop_null = df.dropna()
```

```python
#Count fields to make sure no null

drop_null.count()
```

```python
drop_null.head()
```

```python
#Keep only IATA, Airport, City, State and Country fields

airport_df = drop_null[["IATA","AIRPORT","CITY","STATE","COUNTRY"]]
airport_df.head()
```

```python
#Read into new csv file

airport_df.to_csv("CleanedAirports.csv",index=False)
```
